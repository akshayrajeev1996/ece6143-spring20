### 3A) Generating 10000 random samples from a multivariate distribution 

```python
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

mean = [0,0,0]
cov = [[1,0,0], [0,1,0], [0,0,1]]
x = np.random.multivariate_normal(mean, cov, (10000))
```

### 3B) Computing the Euclidean Norms followed by its average and standard deviation

```python
import statistics 
norms = []
for i in range(0,10000):
  norms.append(np.linalg.norm(x[i]))

hist, bin_edges = np.histogram(norms)
plt.hist(norms, bins = bin_edges)
plt.show()

print("Mean of norms:", statistics.mean(norms))
print("Standard Deviation of norms:", statistics.pstdev(norms))
```

### 3C) Increasing the Dimension scale 

```python
mean_d = []
cov_d = []
xmean_list = []
xsdev_list = []
for j in [50,100,200,500,1000]:
  mean_d = [0] * j
  cov_d = np.identity(j)
  x_d = np.random.multivariate_normal(mean_d, cov_d, (10000))
  norms_d = []
  for i in range(0,10000):
    norms_d.append(np.linalg.norm(x_d[i]))
  hist, bin_edges = np.histogram(norms_d)
  plt.hist(norms_d, bins = bin_edges)
  plt.title("D is " + str(j))
  plt.show()
  print("Mean of norms for d", j, "is:", statistics.mean(norms_d))
  xmean_list.append(statistics.mean(norms_d))
  print("Standard Deviation of norms for d", j, "is:", statistics.pstdev(norms_d))
  xsdev_list.append(statistics.pstdev(norms_d))
```

### Plotting variation in average and standard deviation against increasing d.



```python
plt.plot(xmean_list, [50,100,200,500,1000])
plt.title("Mean Vs Dimension")
plt.xlabel("Means")
plt.ylabel("Dimensions")
plt.show()

plt.plot(xsdev_list, [50,100,200,500,1000], 'o')
plt.title("Standard Deviation vs Dimension")
plt.xlabel("Standard Deviations")
plt.ylabel("Dimensions")
plt.axis([0.5, 1, 0, 1100])
plt.show()
```

### 3D)Mean and Dimension show a positive linear relationship while Standard Deviation tends to remain constant with increasing Dimensions

### 4B) Computing the tf-idf values in all documents

```python
from sklearn.feature_extraction.text import TfidfTransformer
import pandas as pd

doc1 = []
with open('d1.txt') as c1:
  doc1.append(''.join(line.strip() for line in c1))

doc1 = np.asarray(doc1)


doc2 = []
with open('d2.txt') as c2:
  doc2.append(''.join(line.strip() for line in c2))

doc2 = np.asarray(doc2)


doc3 = []
with open('d3.txt') as c3:
  doc3.append(''.join(line.strip() for line in c3))

doc3 = np.asarray(doc3)


doc4 = []
with open('d4.txt') as c4:
  doc4.append(''.join(line.strip() for line in c4))

doc4 = np.asarray(doc4)


doc5 = []
with open('d5.txt') as c5:
  doc5.append(''.join(line.strip() for line in c5))

doc5 = np.asarray(doc5)


docq = []
with open('d_query.txt') as cq:
  docq.append(''.join(line.strip() for line in cq))

docq = np.asarray(docq)

docs = np.append(doc1, doc2)
docs = np.append(docs, doc3)
docs = np.append(docs, doc4)
docs = np.append(docs, doc5)
docs = np.append(docs, docq)


cv = CountVectorizer()
word_count_vector = cv.fit_transform(docs)
word_count_vector.shape

tfidf_transformer = TfidfTransformer(smooth_idf = True, use_idf = True)
tfidf_transformer.fit(word_count_vector)

df_idf = pd.DataFrame(tfidf_transformer.idf_, index = cv.get_feature_names(), columns = ["idf_weights"])

df_idf.sort_values(by = ['idf_weights'])
```

```python
count_vector = cv.transform(docs)

tf_idf_vector = tfidf_transformer.transform(count_vector)

feature_names = cv.get_feature_names()

first_document_vector = tf_idf_vector[0]

print("These are the scores of Document 1: ")
df = pd.DataFrame(first_document_vector.T.todense(), index = feature_names, columns = ["tfidf"])
df.sort_values(by=["tfidf"], ascending = False)
```

## 4C) Calculating the Cosine similarity

```python
from nltk.corpus import stopwords
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

stopWords = stopwords.words('english')

doc1 = str(doc1)
doc2 = str(doc2)
doc3 = str(doc3)
doc4 = str(doc4)
doc5 = str(doc5)
docq = str(docq)

train_string = docq

train_set = [train_string, doc1, doc2, doc3, doc4, doc5]

tfidf_vectorizer = TfidfVectorizer(stop_words = stopWords)

tfidf_matrix_train = tfidf_vectorizer.fit_transform(train_set)

print("Similarity Score = ", cosine_similarity(tfidf_matrix_train[0:1], tfidf_matrix_train))
```
